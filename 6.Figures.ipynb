```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/pradnyajadhav/Desktop/Schizophrenia_Chronopsychosis/2_parts_features(1).csv')

patient_data = df[df['Filename'] == 'patient_1']
control_data = df[df['Filename'] == 'control_5']
x = range(1, 14)  

y_patient = patient_data['part1_mean'].values
y_control = control_data['part1_mean'].values

plt.figure(figsize=(10, 5))
plt.plot(x, y_patient, label='Patient', marker='o', color='red')
plt.plot(x, y_control, label='Control', marker='o',color='blue')
plt.xlabel('Days')
plt.ylabel('Mean of activity data')
plt.title('Mean Values of activity data for Patient and Control')
plt.grid(True)
plt.legend()

day_labels = [f'Day {day}' for day in x]
plt.xticks(x, day_labels, color='black')

plt.show()
```

```python
import matplotlib.pyplot as plt


temporal_patterns = ["12 Parts", "8 parts", "6 Parts", "4 Parts", "3 Parts", "2 Parts", "Full Day", "All Days"]
lightgbm_auc_roc = [0.98, 0.97, 0.98, 0.97, 0.97, 0.97, 0.95, 0.93]
xgboost_auc_roc = [0.97, 0.98, 0.97, 0.97, 0.97, 0.97, 0.95, 0.92]

plt.figure(figsize=(10, 6))
plt.plot(temporal_patterns, lightgbm_auc_roc, marker='o', markersize=10, linestyle='-', label='LightGBM', alpha=0.7)
plt.plot(temporal_patterns, xgboost_auc_roc, marker='s', linestyle='--', label='XGBoost', alpha=0.7)

plt.title('AUC ROC for LightGBM and XGBoost Across Temporal Patterns')
plt.xlabel('Temporal Patterns')
plt.ylabel('AUC ROC')
plt.legend()

plt.grid(True)
plt.xticks(rotation=45)  
plt.show()
```
